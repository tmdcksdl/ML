```python
import pandas as pd

red_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-red.csv'
white_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-white.csv'

red_wine = pd.read_csv(red_url, sep=";")
white_wine = pd.read_csv(white_url, sep=";")
```

```python
red_wine.head()
```

```python
white_wine.head()
```

### white_wine의 column 종류

```python
white_wine.columns
```

```python
red_wine['color'] = 1.0
white_wine['color'] = 0.0

wine = pd.concat([red_wine, white_wine])
wine.info()
```

```python
wine.head()
```

```python
wine['quality'].unique()
```

```python
import plotly.express as px

fig = px.histogram(wine, x='quality')
fig.show()
```

```python
fig = px.histogram(wine, x="quality", color='color')
fig.show()
```

## 레드 와인 / 화이트 와인 분류기

```python
X = wine.drop(['color'], axis=1)
y = wine['color']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

np.unique(y_train, return_counts=True)
```

## 훈련용과 테스트용이 어느 정도로 구분되었는지 확인

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Histogram(x=X_train['quality'], name='Train'))
fig.add_trace(go.Histogram(x=X_test['quality'], name='Test'))

fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig.show()
```

## 결정나무 훈련

```python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=17)
wine_tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

## 데이터 전처리 - MinMaxScaler와 StandardScaler

```python
fig = go.Figure()

fig.add_trace(go.Box(y=X['fixed acidity'], name='fixed acidity'))
fig.add_trace(go.Box(y=X['chlorides'], name='chlorides'))
fig.add_trace(go.Box(y=X['quality'], name='quality'))

fig.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

MMS = MinMaxScaler()
SS = StandardScaler()

SS.fit(X)
MMS.fit(X)

X_ss = SS.transform(X)
X_mms = MMS.transform(X)
```

```python
X_ss_pd = pd.DataFrame(X_ss, columns=X.columns)
X_mms_pd = pd.DataFrame(X_mms, columns=X.columns)
```

```python
def px_box(target_pd):

    fig = go.Figure()

    fig.add_trace(go.Box(y=target_pd['fixed acidity'], name='fixed acidity'))
    fig.add_trace(go.Box(y=target_pd['chlorides'], name='chlorides'))
    fig.add_trace(go.Box(y=target_pd['quality'], name='quality'))

    fig.show()
```

```python
px_box(X_mms_pd)
```

```python
px_box(X_ss_pd)
```

## MinMaxScaler를 적용해서 학습

```python
X_train, X_test, y_train, y_test = train_test_split(X_mms_pd, y, test_size=0.2, random_state=17)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=17)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

## StandardScaler를 적용해서 학습

```python
X_train, X_test, y_train, y_test = train_test_split(X_ss_pd, y, test_size=0.2, random_state=17)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=17)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

## 레드 와인과 화이트 와인을 구분하는 중요 특성

## 와인 맛에 대한 분류 - 이진 분류

```python
wine['taste'] = [1. if grade>5 else 0. for grade in wine['quality']]
wine.info()
```

```python
X = wine.drop(['taste'], axis=1)
y = wine['taste']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=17)
wine_tree.fit(X_train, y_train)
```

```python
y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

```python
import matplotlib.pyplot as plt
import sklearn.tree as tree

plt.figure(figsize=(12,8))
tree.plot_tree(wine_tree, feature_names=list(X.columns));
```
